```python
import json
import requests
import websocket
```

```python
import random, time
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data.csv')
```

```python
df.head()
```

```python
dff = np.array(df)
```

```python
ws = websocket.WebSocket()
ws.connect('ws://localhost:8000/ws/polData/')
```

```python
for i in range(700):
    time.sleep(random.uniform(0,0.3))
    r =dff[i]
    ws.send(json.dumps({'ride_id':r[0],'user_id':r[1],'vehicle_id':r[2],'package_id':r[3],'travel_type_id':r[4],'from_area_id':r[5],'to_area_id':r[6],'from_city_id':r[7],'to_city_id':r[8],'from_date':r[9],'to_date':r[10],'online_booking':r[11],'mobile_site_booking':r[12],'booking_created':r[13],'from_lat':r[14],'from_long':r[15],'to_lat':r[16],'to_long':r[17],'Car_Cancellation':r[18]}))
    res = ws.recv_frame()
    if i%50==0:
        print("Acknowledgement received",res," ",i)
```
